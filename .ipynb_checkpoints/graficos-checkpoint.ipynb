```python
import matplotlib.pyplot as plt
import pandas as pd

# Ler o arquivo CSV
dados = pd.read_csv('USD_BRL_hist.csv')

# Converter a coluna de datas para o formato datetime
dados['Date'] = pd.to_datetime(dados['Date'], format="%d.%m.%Y")

# Definir os intervalos de tempo desejados
data_inicio = '2010-01-01'
data_fim = '2019-12-31'

# Filtrar os dados dentro do intervalo de tempo
dados_filtrados = dados[(dados['Date'] >= data_inicio) & (dados['Date'] <= data_fim)]

# Agrupar os dados por mês e calcular a média do valor do dólar
dados_agrupados = dados_filtrados.groupby(dados_filtrados['Date'].dt.to_period('M')).mean()

# Acessar os dados agrupados
meses = dados_agrupados.index.to_timestamp()
valores_dolar = dados_agrupados['USD_BRL']

# Criação do gráfico de linhas
plt.plot(meses, valores_dolar)

# Personalização do gráfico
plt.xlabel('Mês')
plt.ylabel('Valor do Dólar em Reais')
plt.title('Variação do Dólar em relação ao Real por Mês')

# Exibe o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Ler o arquivo CSV
dados = pd.read_csv('USD_BRL_hist.csv')

# Converter a coluna de datas para o formato datetime
dados['Date'] = pd.to_datetime(dados['Date'], format="%d.%m.%Y")

# Definir os intervalos de tempo desejados
data_inicio = '2018-01-01'
data_fim = '2019-12-31'

# Filtrar os dados dentro do intervalo de tempo
dados_filtrados = dados[(dados['Date'] >= data_inicio) & (dados['Date'] <= data_fim)]

# Acessar os dados filtrados
datas = dados_filtrados['Date']
valores_dolar = dados_filtrados['USD_BRL']

# Encontrar os pontos mais altos do dólar
pontos_altos = dados_filtrados[valores_dolar == valores_dolar.max()]
pontos_medio = dados_filtrados[valores_dolar == valores_dolar.min()]

# Criação do gráfico de linhas
plt.plot(datas, valores_dolar, label='Cotação do Dólar')

# Adicionar os pontos mais altos
plt.scatter(pontos_altos['Date'], pontos_altos['USD_BRL'], color='red', label='Ponto mais alto')
plt.scatter(pontos_medio['Date'], pontos_medio['USD_BRL'], color='red', label='Ponto mais alto')

# Personalização do gráfico
plt.xlabel('Date')
plt.ylabel('Valor do Dólar em Reais')
plt.title('Variação do Dólar em relação ao Real\nGráfico de Linhas')
plt.legend()  # Mostra as legendas

# Exibe o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Ler o arquivo CSV
dados = pd.read_csv('USD_BRL_hist.csv')

# Converter a coluna de datas para o formato datetime
dados['Date'] = pd.to_datetime(dados['Date'], format="%d.%m.%Y")

# Definir os intervalos de tempo desejados
data_inicio = '2019-01-01'
data_fim = '2019-12-31'

# Filtrar os dados dentro do intervalo de tempo
dados_filtrados = dados[(dados['Date'] >= data_inicio) & (dados['Date'] <= data_fim)]

# Acessar os dados filtrados
datas = dados_filtrados['Date']
valores_dolar = dados_filtrados['USD_BRL']

# Encontrar os pontos mais altos e a média do dólar
ponto_alto = dados_filtrados[valores_dolar == valores_dolar.max()]
media_dolar = valores_dolar.mean()

# Criação do gráfico de linhas
plt.plot(datas, valores_dolar, label='Cotação do Dólar')

# Adicionar o ponto mais alto
plt.scatter(ponto_alto['Date'], ponto_alto['USD_BRL'], color='red', label='Ponto mais alto')

# Adicionar a média do dólar
plt.axhline(media_dolar, color='green', linestyle='--', label='Média do Dólar')

# Personalização do gráfico
plt.xlabel('Date')
plt.ylabel('Valor do Dólar em Reais')
plt.title('Variação do Dólar em relação ao Real\nGráfico de Linhas')
plt.legend()  # Mostra as legendas

# Exibe o gráfico
plt.show()
```

```python
from dateutil.relativedelta import relativedelta
import matplotlib.pyplot as plt
import pandas as pd

# Ler o arquivo CSV
dados = pd.read_csv('USD_BRL_hist.csv')

# Converter a coluna de datas para o formato datetime
dados['Date'] = pd.to_datetime(dados['Date'], format="%d.%m.%Y")

# Definir os intervalos de tempo desejados
data_inicio = '2018-01-01'
data_fim = '2019-12-31'

# Filtrar os dados dentro do intervalo de tempo
dados_filtrados = dados[(dados['Date'] >= data_inicio) & (dados['Date'] <= data_fim)]

# Agrupar os dados por mês e calcular a média mensal
dados_mensais = dados_filtrados.resample('M', on='Date').mean()

# Acessar os dados filtrados
datas = dados_filtrados['Date']
valores_dolar = dados_filtrados['USD_BRL']

# Encontrar os pontos mais altos e mais baixos do dólar
ponto_alto = dados_filtrados[valores_dolar == valores_dolar.max()]
ponto_baixo = dados_filtrados[valores_dolar == valores_dolar.min()]

# Criação do gráfico de linhas
plt.plot(datas, valores_dolar, label='Cotação do Dólar')
plt.plot(dados_mensais.index, dados_mensais['USD_BRL'], color='green', label='Média Mensal')

# Adicionar os pontos mais altos e mais baixos
plt.scatter(ponto_alto['Date'], ponto_alto['USD_BRL'], color='red', label='Ponto mais alto ' + str(ponto_alto.iloc[0]["USD_BRL"]))
plt.scatter(ponto_baixo['Date'], ponto_baixo['USD_BRL'], color='blue', label='Ponto mais baixo '+ str(ponto_baixo.iloc[0]["USD_BRL"]))

# Personalização do gráfico
plt.xlabel('Date')
plt.ylabel('Valor do Dólar em Reais')
plt.title('Variação do Dólar em relação ao Real')
plt.legend()  # Mostra as legendas

# Exibe o gráfico
plt.show()
```

```python
from dateutil.relativedelta import relativedelta
import matplotlib.pyplot as plt
import pandas as pd

# Ler o arquivo CSV
dados = pd.read_csv('USD_BRL_hist.csv')

# Converter a coluna de datas para o formato datetime
dados['Date'] = pd.to_datetime(dados['Date'], format="%d.%m.%Y")

# Definir os intervalos de tempo desejados
data_inicio = '2018-01-01'
data_fim = '2019-12-31'

# Filtrar os dados dentro do intervalo de tempo
dados_filtrados = dados[(dados['Date'] >= data_inicio) & (dados['Date'] <= data_fim)]

# Agrupar os dados por mês e calcular a média mensal
dados_mensais = dados_filtrados.resample('M', on='Date').mean()

# Acessar os dados filtrados
datas = dados_filtrados['Date']
valores_dolar = dados_filtrados['USD_BRL']

# Encontrar os pontos mais altos e mais baixos do dólar
ponto_alto = dados_filtrados[valores_dolar == valores_dolar.max()]
ponto_baixo = dados_filtrados[valores_dolar == valores_dolar.min()]

# Criação do gráfico de linhas
plt.plot(datas, valores_dolar, label='Cotação do Dólar')
plt.plot(dados_mensais.index, dados_mensais['USD_BRL'], color='green', label='Média Mensal')

# Adicionar os pontos mais altos e mais baixos
plt.scatter(ponto_alto['Date'], ponto_alto['USD_BRL'], color='red', label='Ponto mais alto ' + str(ponto_alto.iloc[0]["USD_BRL"]))
plt.scatter(ponto_baixo['Date'], ponto_baixo['USD_BRL'], color='blue', label='Ponto mais baixo '+ str(ponto_baixo.iloc[0]["USD_BRL"]))

# Adicionar valor da média mensal
for data, valor in zip(dados_mensais.index, dados_mensais['USD_BRL']):
    plt.annotate(f'{valor:.2f}', (data, valor), textcoords="offset points", xytext=(0,5), ha='center')

# Personalização do gráfico
plt.xlabel('Date')
plt.ylabel('Valor do Dólar em Reais')
plt.title('Variação do Dólar em relação ao Real')
plt.legend()  # Mostra as legendas

# Exibe o gráfico
plt.show()
```

```python
from datetime import datetime
from dateutil.relativedelta import relativedelta
import matplotlib.pyplot as plt
import pandas as pd

# Ler o arquivo CSV
dados = pd.read_csv('USD_BRL_hist.csv')

# Converter a coluna de datas para o formato datetime
dados['Date'] = pd.to_datetime(dados['Date'], format="%d.%m.%Y")

# Definir os intervalos de tempo desejados
data_inicio = '2018-01-01'
data_fim = '2018-12-31'

# Filtrar os dados dentro do intervalo de tempo
dados_filtrados = dados[(dados['Date'] >= data_inicio) & (dados['Date'] <= data_fim)]


# Agrupar os dados por mês e calcular a média mensal
dados_mensais = dados_filtrados.resample('M', on='Date').mean()

meses = relativedelta( datetime.strptime(data_fim, '%Y-%m-%d'),  datetime.strptime(data_inicio, '%Y-%m-%d')).months
soma_media = 0

for valor in dados_mensais.USD_BRL:
    soma_media += valor
media = (soma_media/meses)


# Acessar os dados filtrados
datas = dados_filtrados['Date']
valores_dolar = dados_filtrados['USD_BRL']

# Encontrar os pontos mais altos e mais baixos do dólar
ponto_alto = dados_filtrados[valores_dolar == valores_dolar.max()]
ponto_baixo = dados_filtrados[valores_dolar == valores_dolar.min()]

# Criação do gráfico de linhas
plt.plot(datas, valores_dolar, label='Cotação do Dólar')
plt.plot(dados_mensais.index, dados_mensais['USD_BRL'], color='green', label='Média Mensal')

# Adicionar os pontos mais altos e mais baixos
plt.scatter(dados_mensais.index, dados_mensais['USD_BRL'], color='green', label='Média Total ('+ str(round(media, 2)) + ')')
plt.scatter(ponto_alto['Date'], ponto_alto['USD_BRL'], color='red', label='Ponto mais alto (' + str(round(ponto_alto.iloc[0]["USD_BRL"], 2)) + ')')
plt.scatter(ponto_baixo['Date'], ponto_baixo['USD_BRL'], color='blue', label='Ponto mais baixo (' + str(round(ponto_baixo.iloc[0]["USD_BRL"], 2)) + ')')

# Personalização do gráfico
plt.xlabel('Date')
plt.ylabel('Valor do Dólar em Reais')
plt.title('Variação do Dólar em relação ao Real')
plt.legend()  # Mostra as legendas

# Exibe o gráfico
plt.show()
```
